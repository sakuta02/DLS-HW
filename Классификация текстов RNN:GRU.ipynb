```python
words = Counter()

for example in tqdm(dataset['train']['text']):
    # Приводим к нижнему регистру и убираем пунктуацию
    prccessed_text = example.lower().translate(str.maketrans('', '', string.punctuation))

    for word in word_tokenize(prccessed_text):
        words[word] += 1


vocab = set(['<unk>', '<bos>', '<eos>', '<pad>'])
counter_threshold = 25

for char, cnt in words.items():
    if cnt > counter_threshold:
        vocab.add(char)

print(f'Размер словаря: {len(vocab)}')

word2ind = {char: i for i, char in enumerate(vocab)}
ind2word = {i: char for char, i in word2ind.items()}
```

```python
class WordDataset:
    def __init__(self, sentences):
        self.data = sentences
        self.unk_id = word2ind['<unk>']
        self.bos_id = word2ind['<bos>']
        self.eos_id = word2ind['<eos>']
        self.pad_id = word2ind['<pad>']

    def __getitem__(self, idx: int) -> List[int]:
        processed_text = self.data[idx]['text'].lower().translate(str.maketrans('', '', string.punctuation))
        tokenized_sentence = [self.bos_id]
        tokenized_sentence += [word2ind.get(word, self.unk_id) for word in word_tokenize(processed_text)] 
        tokenized_sentence += [self.eos_id]

        train_sample = {
            "text": tokenized_sentence,
            "label": self.data[idx]['label']
        }

        return train_sample

    def __len__(self) -> int:
        return len(self.data)


def collate_fn_with_padding(input_batch: List[List[int]], pad_id=word2ind['<pad>'], max_len=256) -> torch.Tensor:
    seq_lens = [len(x['text']) for x in input_batch]
    max_seq_len = min(max(seq_lens), max_len)

    new_batch = []
    for sequence in input_batch:
        sequence['text'] = sequence['text'][:max_seq_len]
        for _ in range(max_seq_len - len(sequence['text'])):
            sequence['text'].append(pad_id)

        new_batch.append(sequence['text'])
    
    sequences = torch.LongTensor(new_batch).to(device)
    labels = torch.LongTensor([x['label'] for x in input_batch]).to(device)

    new_batch = {
        'input_ids': sequences,
        'label': labels
    }

    return new_batch
```

```python
train_dataset = WordDataset(dataset['train'])

np.random.seed(42)
idx = np.random.choice(np.arange(len(dataset['test'])), 5000)
eval_dataset = WordDataset(dataset['test'].select(idx))

batch_size = 32
train_dataloader = DataLoader(train_dataset, shuffle=True, collate_fn=collate_fn_with_padding, batch_size=batch_size)

eval_dataloader = DataLoader(eval_dataset, shuffle=False, collate_fn=collate_fn_with_padding, batch_size=batch_size)
```

## Постановка задачи
Ваша задача -- получить максимальное возможное accuracy на `eval_dataloader`. Ниже приведена функция, которую вам необходимо запустить для обученной модели, чтобы вычислить качество её работы.

```python
def evaluate(model, eval_dataloader) -> float:
    """
    Calculate accuracy on validation dataloader.
    """

    predictions = []
    target = []
    with torch.no_grad():
        for batch in eval_dataloader:
            logits = model(batch['input_ids'])
            predictions.append(logits.argmax(dim=1))
            target.append(batch['label'])
    
    predictions = torch.cat(predictions)
    target = torch.cat(target)
    accuracy = (predictions == target).float().mean().item()

    return accuracy
```

```python
from torch import nn


class RNNClassifier(nn.Module):
    def __init__(self, vocab_size, hidden_size, num_classes=4):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, hidden_size)

        self.rnn = nn.RNN(hidden_size, hidden_size, batch_first=True)

        self.linear = nn.Linear(hidden_size, hidden_size)
        self.non_linear = nn.Tanh()
        self.dropout = nn.Dropout(p=0.1)

        self.projection = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        x = self.embedding(x)
        x, _ = self.rnn(x)
        x = x.mean(dim=1)

        x = self.linear(x)
        x = self.non_linear(x)
        x = self.dropout(x)

        x = self.projection(x)

        return x
```

```python
from torch.optim import Adam
from torch.nn import CrossEntropyLoss

model = RNNClassifier(len(vocab), 256).to(device)
optimizer = Adam(model.parameters())
criterion = CrossEntropyLoss(ignore_index=word2ind['<pad>'])
```

```python
model.train()
losses = []
num_epochs = 5
eval_steps = len(train_dataloader) // 2
acc = []

for _ in tqdm(range(num_epochs)):
    epochs_loss = []
    for i, batch in enumerate(tqdm(train_dataloader)):
        optimizer.zero_grad()
        logits = model(batch['input_ids'])
        loss = criterion(logits, batch['label'])
        loss.backward()
        optimizer.step()
        
        epochs_loss.append(loss.item())
        if i % eval_steps == 0:
            model.eval()
            acc.append(evaluate(model, eval_dataloader))
            model.train()
    
    losses.append(sum(epochs_loss) / len(epochs_loss))
```

# Эксперименты

```python
def train(model, optimizer, criterion, train_dataloader, eval_dataloader):
    model.train()
    losses = []
    num_epochs = 5
    eval_steps = len(train_dataloader) // 2
    acc = []
    
    for _ in tqdm(range(num_epochs)):
        epochs_loss = []
        for i, batch in enumerate(tqdm(train_dataloader)):
            optimizer.zero_grad()
            logits = model(batch['input_ids'])
            loss = criterion(logits, batch['label'])
            loss.backward()
            optimizer.step()
            
            epochs_loss.append(loss.item())
            if i % eval_steps == 0:
                model.eval()
                acc.append(evaluate(model, eval_dataloader))
                model.train()
        
        losses.append(sum(epochs_loss) / len(epochs_loss))
    return losses, acc
```

```python
def print_data(losses, acc):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

    ax1.plot(np.arange(len(losses)), losses, color='red')
    ax1.set_title('Loss')
    ax1.set_xlabel("epoch")

    ax2.plot(np.arange(len(acc)), acc, color='red')
    ax2.set_title('Accuracy')
    ax2.set_xlabel("Measurement")

    plt.show()
    
```

```python
class GRUClassifier(nn.Module): # Посмотрим на улучшение при помощи GRU и max пуллинга
    def __init__(self, vocab_size, hidden_size, num_classes=4):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, hidden_size)

        self.rnn = nn.GRU(hidden_size, hidden_size, batch_first=True)

        self.linear = nn.Linear(hidden_size, hidden_size)
        self.non_linear = nn.Tanh()
        self.dropout = nn.Dropout(p=0.1)

        self.projection = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        x = self.embedding(x)
        x, _ = self.rnn(x)
        x = x.max(dim=1)[0]

        x = self.linear(x)
        x = self.non_linear(x)
        x = self.dropout(x)

        x = self.projection(x)

        return x

model_gru = GRUClassifier(len(vocab), 256).to(device)
optimizer_gru = Adam(model_gru.parameters())
criterion_gru = CrossEntropyLoss(ignore_index=word2ind['<pad>'])

losses_gru, acc_gru = train(model_gru, optimizer_gru, criterion_gru, train_dataloader, eval_dataloader)
```

```python
print_data(losses_gru, acc_gru)
```

```python
class GRU2Classifier(nn.Module): # Посмотрим на улучшение при помощи 2 слоев GRU + 1 fc слоя
    def __init__(self, vocab_size, hidden_size, num_classes=4):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, hidden_size)

        self.rnn_1 = nn.GRU(hidden_size, hidden_size, num_layers=2, batch_first=True)


        self.linear_1 = nn.Linear(hidden_size, hidden_size)
        self.non_linear_1 = nn.ReLU()
        self.dropout_1 = nn.Dropout(p=0.1)

        self.linear_2 = nn.Linear(hidden_size, hidden_size)
        self.non_linear_2 = nn.ReLU()
        self.dropout_2 = nn.Dropout(p=0.1)

        self.projection = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        x = self.embedding(x)
        x, _ = self.rnn_1(x)
        x = x.max(dim=1)[0]
        

        x = self.linear_1(x)
        x = self.non_linear_1(x)
        x = self.dropout_1(x)

        x = self.linear_2(x)
        x = self.non_linear_2(x)
        x = self.dropout_2(x)
        

        x = self.projection(x)

        return x

model_gru2 = GRU2Classifier(len(vocab), 256).to(device)
optimizer_gru2 = Adam(model_gru2.parameters())
criterion_gru2 = CrossEntropyLoss(ignore_index=word2ind['<pad>'])

losses_gru2, acc_gru2 = train(model_gru2, optimizer_gru2, criterion_gru2, train_dataloader, eval_dataloader)
```

```python
print_data(losses_gru2, acc_gru2)
```
